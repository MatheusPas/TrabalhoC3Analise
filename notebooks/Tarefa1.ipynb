### Função de conexão com o servidor e buscar o token do meu user na API 

```python
def create_con(user, password):
    response = requests.post('http://localhost:8000/auth/login', json={
        "username": user,
        "password": password
    })
    token = response.json()['access_token']
    return {'Authorization': f'Bearer {token}'}
```

```python
create_con('matheus_root', 'server@Database')
```
